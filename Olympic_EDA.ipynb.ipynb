```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

age_outliers = df[(df["Age"] < lower) | (df["Age"] > upper)]
age_outliers.shape
```

## Outlier Detection in Age

Using the IQR method, 11,062 records were identified as potential outliers in the Age column. However, since Olympic data may naturally include very young or older athletes depending on the sport, these values were considered valid and were not removed.

```python
df["Age"].min(), df["Age"].max()
```

## Age Range Check

The minimum and maximum age values were examined to understand the range of athlete ages. The values appear realistic for Olympic participation, so extreme values were retained.

```python
Q1 = df["Height"].quantile(0.25)
Q3 = df["Height"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

height_outliers = df[(df["Height"] < lower) | (df["Height"] > upper)]
height_outliers.shape
```

## Outlier Detection in Height

```python
Q1 = df["Weight"].quantile(0.25)
Q3 = df["Weight"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

weight_outliers = df[(df["Weight"] < lower) | (df["Weight"] > upper)]
weight_outliers.shape
```
